```python
import tensorflow as tf
```

```python
scaler = tf.constant(5)
print(tf.rank(scaler))
```

```python
vector = tf.constant([1,23,3,4])
print(tf.rank(vector))
```

```python
matrix = tf.constant([[1,23],[23,432]])
print(tf.rank(matrix))
```

```python
tensor_3d = tf.constant([[[12,2],[213,32]],[[23,45],[43,43]]])
print(tf.rank(tensor_3d))
```

```python
A = tf.constant([1,2,3,4,5])
B=tf.constant([3,4,56,6,343])
```

```python
sum_result = tf.add(A,B)
print(sum_result)
```

```python
sub_result = tf.subtract(A,B)
print(sub_result)
```

```python
elem_mul = tf.multiply(A,B)
print(elem_mul)
```

```python
A_reshaped = tf.reshape(A,(1,-1))
B_reshaped = tf.reshape(B,(-1,1))
```

```python
mat_mul = tf.matmul(A_reshaped,B_reshaped)
print(mat_mul)
```
